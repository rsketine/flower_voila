```python
#upload pic
from ipywidgets import widgets
from fastai.vision.all import *
from fastai.vision.widgets import *
from fastbook import *

btn_upload = widgets.FileUpload()
btn_upload
```

```python
#we cannot upload via widget now so faking it with this upload
path = Path('flowers/rose')
path.absolute()
f = f'{path}/rose.jpg'
print(f)
btn_upload = SimpleNamespace(data = [f])
img = PILImage.create(btn_upload.data[-1])
#img.to_thumb(128, 128)
```

```python
#display the widget
out_pl = widgets.Output()
out_pl.clear_output()
with out_pl: display(img.to_thumb(128, 128))
out_pl
```

```python
model_path = Path('export.pkl')
learn_inf = load_learner(model_path)
pred, pred_index, probs = learn_inf.predict(img)
```

```python
#hide_output
lbl_predict = widgets.Label()
lbl_predict.value = f'Prediction: {pred}, probability: {probs[pred_index]:.4f}'
lbl_predict
```

```python
btn_run = widgets.Button(description="Classify")
btn_run
```

```python
#Click event code to run
def on_click_classify(change):
    img = PILImage.create(button_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128 * 128))
    pred, pred_index, probs=learn_inf.predict(img)
    lbl_predict.value = f'Prediction: {pred} Probability: {probs[pred_index]:.4f}'

btn_run.on_click(on_click_classify)
```

```python
#hide_output
#putting back btn_upload for a widget for fileUpload
btn_upload = widgets.FileUpload()
```

```python
#hide_output
#set VBox
VBox([widgets.Label('Select your flower'), btn_upload, btn_run, out_pl, lbl_predict])
```
